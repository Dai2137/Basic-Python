```python
df = pd.read_csv('C:\PlayGround\Basic-Python\colab\quizzes\matplotlib_quiz\hist_data.csv')
a,b = df['aデータ'], df['bデータ']
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度（℃）')
plt.ylabel('度数')
plt.hist(a, color = "orange", bins = 30, label = "aデータ", alpha = 0.5)
plt.hist(b, color = "skyblue", bins = 30, label = "bデータ", alpha = 0.5)
plt.legend(loc = 1)
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
df = pd.read_csv("C:\PlayGround\Basic-Python\colab\quizzes\matplotlib_quiz\scatter_data.csv")
c0 = df[df["class"] == 0]
c1 = df[df["class"] == 1]
c2 = df[df["class"] == 2]
plt.scatter(c0["length (cm)"],c0["width (cm)"],color = "r", label = "class0")
plt.scatter(c1["length (cm)"],c1["width (cm)"],color = "b", label = "class1")
plt.scatter(c2["length (cm)"],c2["width (cm)"],color = "g", label = "class2")
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width(cm)')
plt.grid(True)
plt.legend(loc = 1)
```

```python
df = pd.read_csv("C:\PlayGround\Basic-Python\colab\quizzes\matplotlib_quiz\line_data.csv")
fig = plt.figure(figsize =(9,12))
ax1 = fig.add_subplot(211)
ax1.set_title('期間ごとの件数の推移')
ax1.set_xlabel('期間')
ax1.set_ylabel('件数')
ax1.grid(True)
ax1.set_yticks(np.arange(105,141,5))
ax1.plot(df['期間'],df['件数'], marker = "o")

ax2 =fig.add_subplot(212)
ax2.set_title('期間ごとの費用の推移')
ax2.set_xlabel('期間')
ax2.set_ylabel('費用（円）')
ax2.grid(True)
ax2.bar(df['期間'],df['費用'])
```
